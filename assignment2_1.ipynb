```python
import os
import torch
import torchaudio
import torchaudio.functional as F
import torchaudio.transforms as T
import math
import timeit
import librosa
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Audio
import resampy

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

def benchmark_resample_functional(
    waveform,
    sample_rate,
    resample_rate,
    lowpass_filter_width=6,
    rolloff=0.99,
    resampling_method="sinc_interp_hann",
    beta=None,
    iters=5,
):
    return (
        timeit.timeit(
            stmt="""
torchaudio.functional.resample(
    waveform,
    sample_rate,
    resample_rate,
    lowpass_filter_width=lowpass_filter_width,
    rolloff=rolloff,
    resampling_method=resampling_method,
    beta=beta,
)
            """,
            setup="import torchaudio",
            number=iters,
            globals=locals(),
        )
        * 1000
        / iters
    )

def benchmark_resample_transforms(
    waveform,
    sample_rate,
    resample_rate,
    lowpass_filter_width=6,
    rolloff=0.99,
    resampling_method="sinc_interp_hann",
    beta=None,
    iters=5,
):

    return (
        timeit.timeit(
            stmt="resampler(waveform)",
            setup="""
import torchaudio
resampler = torchaudio.transforms.Resample(
    sample_rate,
    resample_rate,
    lowpass_filter_width=lowpass_filter_width,
    rolloff=rolloff,
    resampling_method=resampling_method,
    dtype=waveform.dtype,
    beta=beta,
)
resampler.to(waveform.device)
            """,
            number=iters,
            globals=locals(),
        )
        * 1000
        / iters
    )
def benchmark_resample_librosa(
    waveform,
    sample_rate,
    resample_rate,
    res_type=None,
    iters=5,
):

    waveform_np = waveform.squeeze().numpy()
    return (
        timeit.timeit(
            stmt="""
librosa.resample(
    waveform_np,
    orig_sr=sample_rate,
    target_sr=resample_rate,
    res_type=res_type,
)
            """,
            setup="import librosa",
            number=iters,
            globals=locals(),
        )
        * 1000
        / iters
    )

def plot_benchmark_results(df, file_name, direction):

    print(f"\n--- {file_name} {direction} 벤치마크 ---")
    print(df.round(2))
    ax = df.plot(kind="bar")
    plt.title(f"Resampling Benchmark for {file_name} ({direction})")
    plt.ylabel("Time Elapsed [ms]")
    plt.xticks(rotation=0, fontsize=10)
    for cont, col, color in zip(ax.containers, df.columns, mcolors.TABLEAU_COLORS):
        label = ["N/A" if v != v else str(v) for v in df[col].round(2)]
        ax.bar_label(cont, labels=label, color=color, fontweight="bold", fontsize="x-small")
    plt.show()

def benchmark_file(file_path, downsample_rate, upsample_rate):

    try:
        waveform, sample_rate = torchaudio.load(file_path)
        file_name = os.path.basename(file_path)


        waveform = waveform.to(torch.float32).cpu()


        times_down, rows_down = [], []
        args_down = (waveform, sample_rate, downsample_rate)

        # sinc 64 zero-crossings
        f_time = benchmark_resample_functional(*args_down, lowpass_filter_width=64)
        t_time = benchmark_resample_transforms(*args_down, lowpass_filter_width=64)
        times_down.append([None, f_time, t_time])
        rows_down.append(f"sinc (width 64)")

        # sinc 6 zero-crossings
        f_time = benchmark_resample_functional(*args_down, lowpass_filter_width=16)
        t_time = benchmark_resample_transforms(*args_down, lowpass_filter_width=16)
        times_down.append([None, f_time, t_time])
        rows_down.append("sinc (width 16)")

        # kaiser best
        kwargs_best = {
            "lowpass_filter_width": 64,
            "rolloff": 0.9475937167399596,
            "resampling_method": "sinc_interp_kaiser",
            "beta": 14.769656459379492,
        }
        lib_time = benchmark_resample_librosa(*args_down, res_type="kaiser_best")
        f_time = benchmark_resample_functional(*args_down, **kwargs_best)
        t_time = benchmark_resample_transforms(*args_down, **kwargs_best)
        times_down.append([lib_time, f_time, t_time])
        rows_down.append("kaiser_best")

        # kaiser fast
        kwargs_fast = {
            "lowpass_filter_width": 16,
            "rolloff": 0.85,
            "resampling_method": "sinc_interp_kaiser",
            "beta": 8.555504641634386,
        }
        lib_time = benchmark_resample_librosa(*args_down, res_type="kaiser_fast")
        f_time = benchmark_resample_functional(*args_down, **kwargs_fast)
        t_time = benchmark_resample_transforms(*args_down, **kwargs_fast)
        times_down.append([lib_time, f_time, t_time])
        rows_down.append("kaiser_fast")

        df_down = pd.DataFrame(times_down, columns=["librosa", "functional", "transforms"], index=rows_down)
        plot_benchmark_results(df_down, file_name, f"downsampling ({sample_rate} -> {downsample_rate})")

        times_up, rows_up = [], []
        args_up = (waveform, sample_rate, upsample_rate)

        # sinc 64 zero-crossings
        f_time = benchmark_resample_functional(*args_up, lowpass_filter_width=64)
        t_time = benchmark_resample_transforms(*args_up, lowpass_filter_width=64)
        times_up.append([None, f_time, t_time])
        rows_up.append(f"sinc (width 64)")

        # sinc 6 zero-crossings
        f_time = benchmark_resample_functional(*args_up, lowpass_filter_width=16)
        t_time = benchmark_resample_transforms(*args_up, lowpass_filter_width=16)
        times_up.append([None, f_time, t_time])
        rows_up.append("sinc (width 16)")

        # kaiser best
        lib_time = benchmark_resample_librosa(*args_up, res_type="kaiser_best")
        f_time = benchmark_resample_functional(*args_up, **kwargs_best)
        t_time = benchmark_resample_transforms(*args_up, **kwargs_best)
        times_up.append([lib_time, f_time, t_time])
        rows_up.append("kaiser_best")

        # kaiser fast
        lib_time = benchmark_resample_librosa(*args_up, res_type="kaiser_fast")
        f_time = benchmark_resample_functional(*args_up, **kwargs_fast)
        t_time = benchmark_resample_transforms(*args_up, **kwargs_fast)
        times_up.append([lib_time, f_time, t_time])
        rows_up.append("kaiser_fast")

        df_up = pd.DataFrame(times_up, columns=["librosa", "functional", "transforms"], index=rows_up)
        plot_benchmark_results(df_up, file_name, f"upsampling ({sample_rate} -> {upsample_rate})")

    except Exception as e:
        print(f"Error processing '{file_path}': {e}")



audio_files = [
    "/content/095522039.m4a",
    "/content/095522040.m4a",
    "/content/095522041.m4a",
    "/content/095522042.m4a"
]

downsample_rate = 16000
upsample_rate = 48000

for file_path in audio_files:
    benchmark_file(file_path, downsample_rate, upsample_rate)
```
